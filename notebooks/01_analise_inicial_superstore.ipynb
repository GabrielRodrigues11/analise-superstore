```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gr√°ficos configurados

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
caminho_arquivo = '../dados/brutos/superstore.csv'

df = pd.read_csv(caminho_arquivo, encoding='latin1')
```

# üöÄ Sele√ß√£o, limpeza e transforma√ß√£o dos dados

```python
df = df.rename(columns={
    'Row ID': 'ID da Linha',
    'Order ID': 'ID do Pedido',
    'Order Date': 'Data do Pedido',
    'Ship Date': 'Data de Envio',
    'Ship Mode': 'Modo de Envio',
    'Customer ID': 'ID do Cliente',
    'Customer Name': 'Nome do Cliente',
    'Segment': 'Segmento',
    'City': 'Cidade',
    'State': 'Estado',
    'Country': 'Pa√≠s',
    'Postal Code': 'CEP',
    'Market': 'Mercado',
    'Region': 'Regi√£o',
    'Product ID': 'ID do Produto',
    'Category': 'Categoria',
    'Sub-Category': 'Subcategoria',
    'Product Name': 'Nome do Produto',
    'Sales': 'Vendas',
    'Quantity': 'Quantidade',
    'Discount': 'Desconto',
    'Profit': 'Lucro',
    'Shipping Cost': 'Custo de Envio',
    'Order Priority': 'Prioridade do Pedido'
})
```

```python
df['CEP'] = df['CEP'].fillna('N√£o informado')
```

```python
df['Data do Pedido'] = pd.to_datetime(df['Data do Pedido'], errors='coerce')
df['Data de Envio'] = pd.to_datetime(df['Data de Envio'], errors='coerce')
```

```python
df['Ano do Pedido'] = df['Data do Pedido'].dt.year
df['M√™s do Pedido'] = df['Data do Pedido'].dt.month
df['Nome do M√™s do Pedido'] = df['Data do Pedido'].dt.strftime('%B')
```

```python
df.head()
```

# üìä An√°lise Explorat√≥ria de Pedidos e Vendas

## üìä 1. Pedidos por Ano

```python
pedidos_por_ano = df['Ano do Pedido'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.barplot(x=pedidos_por_ano.index, y=pedidos_por_ano.values, palette='Blues_d')
plt.title('Quantidade de Pedidos por Ano')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Pedidos')
plt.tight_layout()
plt.savefig('../outputs/graficos/pedidios_por_ano.png', dpi=300)
plt.show()
```

## üìä 2. Pedidos por m√™s (agregados de todos os anos)

```python
meses_traducao = {
    'January': 'Janeiro', 'February': 'Fevereiro', 'March': 'Mar√ßo',
    'April': 'Abril', 'May': 'Maio', 'June': 'Junho',
    'July': 'Julho', 'August': 'Agosto', 'September': 'Setembro',
    'October': 'Outubro', 'November': 'Novembro', 'December': 'Dezembro'
}
```

```python
df['Nome do M√™s do Pedido'] = df['Nome do M√™s do Pedido'].map(meses_traducao)

pedidos_por_mes = df['Nome do M√™s do Pedido'].value_counts()

meses_ordem = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho',
               'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

pedidos_por_mes = pedidos_por_mes.reindex(meses_ordem, fill_value=0)
```

```python
sns.barplot(x=pedidos_por_mes.index, y=pedidos_por_mes.values, palette='viridis')
plt.title('Pedidos por M√™s (Agregado)')
plt.xlabel('M√™s')
plt.ylabel('N√∫mero de Pedidos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/graficos/pedidos_por_mes.png', dpi=300)
plt.show()
```

## üßç 3. Segmentos de Clientes

```python
segmentos = df['Segmento'].value_counts()

sns.barplot(x=segmentos.index, y=segmentos.values, palette='pastel')
plt.title('Distribui√ß√£o dos Segmentos de Cliente')
plt.xlabel('Segmento')
plt.ylabel('N√∫mero de Pedidos')
plt.tight_layout()
plt.savefig('../outputs/graficos/segmentos_clientes.png', dpi=300)
plt.show()
```

## üöö 4. Modos de Envio Utilizados

```python
modos_envio = df['Modo de Envio'].value_counts()

sns.barplot(x=modos_envio.index, y=modos_envio.values, palette='coolwarm')
plt.title('Modos de Envio Utilizados')
plt.xlabel('Modo de Envio')
plt.ylabel('N√∫mero de Pedidos')
plt.tight_layout()
plt.savefig('../outputs/graficos/modos_envio.png', dpi=300)
plt.show()
```

## üó∫Ô∏è 5. Estados ou cidades com mais pedidos

```python
top_estados = df['Estado'].value_counts().head(10)

sns.barplot(x=top_estados.values, y=top_estados.index, palette='mako')
plt.title('Top 10 Estados com Mais Pedidos')
plt.xlabel('N√∫mero de Pedidos')
plt.ylabel('Estado')
plt.tight_layout()
plt.savefig('../outputs/graficos/top_estados_pedidos.png', dpi=300)
plt.show()
```

## üìä 6. Vendas Totais por Categoria

```python
vendas_por_categoria = df.groupby('Categoria')['Vendas'].sum().sort_values(ascending=False)

sns.barplot(x=vendas_por_categoria.values, y=vendas_por_categoria.index, palette='rocket')
plt.title('Vendas Totais por Categoria')
plt.xlabel('Vendas (R$)')
plt.ylabel('Categoria')
plt.tight_layout()
plt.savefig('../outputs/graficos/vendas_por_categoria.png', dpi=300)
plt.show()
```

## üìà 7. Lucro Total por Regi√£o

```python
lucro_por_regiao = df.groupby('Regi√£o')['Lucro'].sum().sort_values(ascending=False)

sns.barplot(x=lucro_por_regiao.values, y=lucro_por_regiao.index, palette='mako')
plt.title('Lucro Total por Regi√£o')
plt.xlabel('Lucro (R$)')
plt.ylabel('Regi√£o')
plt.tight_layout()
plt.savefig('../outputs/graficos/lucro_por_regiao.png', dpi=300)
plt.show()
```

## üìä 8. Vendas Totais por Estado

```python
vendas_por_estado = df.groupby('Estado')['Vendas'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=vendas_por_estado.values, y=vendas_por_estado.index, palette='coolwarm')
plt.title('Top 10 Estados por Vendas Totais')
plt.xlabel('Vendas (R$)')
plt.ylabel('Estado')
plt.tight_layout()
plt.savefig('../outputs/graficos/vendas_por_estado.png', dpi=300)
plt.show()
```

## üìà 9. Lucro Total por Cidade

```python
lucro_por_cidade = df.groupby('Cidade')['Lucro'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=lucro_por_cidade.values, y=lucro_por_cidade.index, palette='viridis')
plt.title('Top 10 Cidades por Lucro Total')
plt.xlabel('Lucro (R$)')
plt.ylabel('Cidade')
plt.tight_layout()
plt.savefig('../outputs/graficos/lucro_por_cidade.png', dpi=300)
plt.show()
```

## üìÖ 10. An√°lise temporal detalhada - vendas ao longo do tempo

```python
df['Ano-M√™s'] = df['Data do Pedido'].dt.to_period('M')

vendas_ano_mes = df.groupby('Ano-M√™s')['Vendas'].sum().reset_index()

vendas_ano_mes['Ano-M√™s'] = vendas_ano_mes['Ano-M√™s'].dt.to_timestamp()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=vendas_ano_mes, x='Ano-M√™s', y='Vendas', marker='o')
plt.title('Evolu√ß√£o das Vendas ao Longo do Tempo')
plt.xlabel('Ano-M√™s')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/graficos/vendas_ao_longo_do_tempo.png', dpi=300)
plt.show()
```

## üìâ 11. Rela√ß√£o entre Desconto e Lucro


```python
plt.figure(figsize=(14,7))
scatter = plt.scatter(
    x=df['Desconto'],
    y=df['Lucro'],
    s=df['Quantidade']*10, 
    c=df['Vendas'],          
    cmap='viridis',
    alpha=0.6,
    edgecolors='w',
    linewidth=0.5
)

plt.colorbar(scatter, label='Vendas (R$)')
plt.title('Rela√ß√£o entre Desconto, Lucro e Quantidade')
plt.xlabel('Desconto')
plt.ylabel('Lucro (R$)')
plt.tight_layout()
plt.savefig('../outputs/graficos/desconto_vs_lucro.png', dpi=300)
plt.show()
```
